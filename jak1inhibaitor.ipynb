```python
import pandas as pd
```

```python
dataset_url = "https://raw.githubusercontent.com/DimBlade/bioinformatic/refs/heads/main/jak1_inhibator_06_bioactivity_data_3class_pIC50_pubchem_fp.csv"
```

```python
dataset = pd.read_csv(dataset_url)
```

```python
dataset
```

```python
X = dataset.drop(('pIC50'), axis=1)
```

```python
X
```

```python
Y = dataset.iloc[:, -1]
Y
```

```python
pip install scikit-learn
```

```python
from sklearn.feature_selection import VarianceThreshold

def remove_low_variance(input_data, threshhold=0.1):
    selection = VarianceThreshold(threshhold)
    selection.fit(input_data)
    return input_data[input_data.columns[selection.get_support(indices=True)]]
X = remove_low_variance(X , threshhold=0.1)
X
```

```python
X.to_csv('descriptor_list.csv' , index = False)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import numpy as np
import pandas as pd

print("NaN in Y:", np.isnan(Y).sum())
print("Inf in Y:", np.isinf(Y).sum())
print("Values > 1e308:", np.sum(np.abs(Y) > 1e308))
```

```python
Y = pd.to_numeric(Y, errors='coerce')   #     NaN
Y = Y.replace([np.inf, -np.inf], np.nan)
Y = Y.fillna(Y.mean())   #  : Y.dropna()
```

```python
model = RandomForestRegressor(n_estimators=500, random_state=42)
model.fit(X, Y)
r2 = model.score(X, Y)
r2
```

```python
Y_pred = model.predict(X)
Y_pred
```

```python
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y, Y_pred))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y, Y_pred))
```
